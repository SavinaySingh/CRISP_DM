## 3.2. Label Encoding

```python
labelencoder = LabelEncoder()
df['gender']=labelencoder.fit_transform(df['gender'])
df['car_model']=labelencoder.fit_transform(df['car_model'])
df['car_segment']=labelencoder.fit_transform(df['car_segment'])
```

# 5. Train Test Split

```python
X=df.drop('Target',axis=1)
y=df.Target
```

```python
assert X.shape==(131337, 14)
assert y.shape==(131337, )
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
assert X_train.shape==(87995, 14)
assert y_train.shape==(87995,)
assert X_test.shape==(43342, 14)
assert y_test.shape==(43342,)
```

# 6. Handling the class imbalance

```python
y_train.value_counts()
```

```python
sm = SMOTE(random_state=42)
```

```python
X_train_sm, y_train_sm = sm.fit_resample(X_train, y_train)
```

```python
y_train_sm.value_counts()
```

# 7. Final Data Visualisation

<p style="font-size:18px;">
    The grouped bar chart below displays different categories of a feature with each set of bars representing a specific category, and each bar within the set indicating the target value associated to that category.
</p>

```python
df2_train=X_train_sm.copy()
df2_train['Target']=y_train_sm
```

```python
df2_train_2=df2_train.copy()
```

```python
labelencoder.fit_transform(df2['gender'])
df2_train_2['gender']=labelencoder.inverse_transform(df2_train_2['gender'])
labelencoder.fit_transform(df2['car_model'])
df2_train_2['car_model']=labelencoder.inverse_transform(df2_train_2['car_model'])
labelencoder.fit_transform(df2['car_segment'])
df2_train_2['car_segment']=labelencoder.inverse_transform(df2_train_2['car_segment'])
```

```python
for col in X:
    df_serv=pd.DataFrame(df2_train_2[df2_train_2.Target==0].groupby(by=col).count().Target)
    df_serv.columns=[ 'Target0']
    df_serv['Target1']=df2_train_2[df2_train_2.Target==1].groupby(by=col).count().Target
    df_serv.plot(kind='bar')
```

# 8. Feature Scaling

```python
# Feature Scaling using Standardization: range [0,1]
scaler = preprocessing.StandardScaler().fit(X_train_sm.append(X_test))
X_train_sm = scaler.transform(X_train_sm)
X_test = scaler.transform(X_test)
```

```python
assert isinstance(X_train_sm, np.ndarray)
assert isinstance(X_test, np.ndarray)
```

# 9. Modelling and Evaluation

```python
# Define the classifiers
logreg = LogisticRegression()
rf = RandomForestClassifier()
svm = SVC()
knn = KNeighborsClassifier()
nb = GaussianNB()
xgb = XGBClassifier(use_label_encoder=False)
```

```python
# Train and test the classifiers
classifiers = {'Logistic Regression': logreg, 'Random Forest': rf, 'SVM': svm, 'KNN': knn, 'Naive Bayes': nb, 'XGBoost':xgb}
accuracies = {}
```

<p style="font-size:18px;">
    A comparative analysis of different classifiers for testing dataset
</p>

```python
for name, clf in classifiers.items():
    clf.fit(X_train_sm, y_train_sm)
    y_pred = clf.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    accuracies[name] = acc

# Plot the bar chart
fig, ax = plt.subplots()
ax.set_xlim([0.6, 1])
ax.barh(list(accuracies.keys()), list(accuracies.values()))
ax.set_title('Comparison of Classifier Accuracies')
ax.set_xlabel('Classifier')
ax.set_ylabel('Accuracy')
plt.xticks(rotation=45)
plt.show()
```

<p style="font-size:18px;">
    A comparative analysis of different classifiers for training dataset
</p>

```python
for name, clf in classifiers.items():
    clf.fit(X_train_sm, y_train_sm)
    y_pred = clf.predict(X_train_sm)
    acc = accuracy_score(y_train_sm, y_pred)
    accuracies[name] = acc

# Plot the bar chart
fig, ax = plt.subplots()
ax.set_xlim([0.6, 1])
ax.barh(list(accuracies.keys()), list(accuracies.values()))
ax.set_title('Comparison of Classifier Accuracies')
ax.set_xlabel('Classifier')
ax.set_ylabel('Accuracy')
plt.xticks(rotation=45)
plt.show()
```

<p style="font-size:18px;">
    It is evident that Random Forest classifier gives the highest accuracy for both testing and training data
</p>

```python
y_pred = rf.predict(X_test)
```

```python
print(classification_report(y_test, y_pred, target_names=['Will not rebuy','Will rebuy']))
```
